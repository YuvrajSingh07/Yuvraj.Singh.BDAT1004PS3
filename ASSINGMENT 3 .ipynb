```python
import pandas as pd

users = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user', sep = '|')
users.head()
```

```python
#maean age by occupation
users.groupby('occupation').age.mean()
```

```python
#male ratio in each occupation
users.groupby('occupation').gender.value_counts()
```

```python
#max age by each occupation
max_age = users.groupby('occupation').age.max()
```

```python
print(max_age)
```

```python
#min age by each occupation
min_age = users.groupby('occupation').age.min()
print(min_age)
```

```python
#meaann age by occupation and gender
users.groupby(['occupation','gender']).age.mean()
```

```python
#percentage of male and females in  
users.groupby('occupation')['gender'].value_counts(normalize=True)*100
```

```python
chipo = pd.read_csv(r'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep ='\t')
chipo
```

```python
#first 10 entries
chipo.head(10)
```

```python
#number of entries
chipo.info()
```

```python
#number of columns
cols = chipo.columns
print(len(cols))
```

```python
#column names
for name in cols:
    print(name)
```

```python
#most ordered item
chipo['item_name'].value_counts()
```

```python
#most ordered in the choice_description
chipo['choice_description'].value_counts()
```

```python
#total items ordered
chipo['quantity'].sum()
```

```python
#different items sold
chipo['item_name'].nunique()
```

```python
#using lambda
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:-1]))
```

```python
chipo.dtypes
```

```python
#total revenue
chipo['revenue'] = chipo['quantity']*chipo['item_price']
chipo['revenue'].sum()
```

```python
#total orders
total_orders = chipo['order_id'].nunique()
total_orders
```

```python
#what is the average revenue amunt per order
chipo['revenue'].sum()/total_orders
```

```python
#QUES 6
import pandas as pd

marriage = pd.read_csv(r'E:\SEMESTER 1\DATA PROGRAMMING\ASSIGNMENT 3\us-marriages-divorces-1867-2014.csv')
marriage.head()
```

```python
marriage.dtypes
marriage.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig, ax = plt.subplots()
plt.plot(marriage['Year'], marriage['Marriages_per_1000'], label = "Marriage")
plt.plot(marriage['Year'], marriage['Divorces_per_1000'], label = "Divorces")
plt.legend()
plt.title("Marriages and divorces per capita")
plt.xlabel("Year")
plt.ylabel("Number per capita")
```

```python
#QUES 7
```

```python
df1900 = marriage[marriage.Year == 1900]
df1950 = marriage[marriage.Year == 1950]
df2000 = marriage[marriage.Year == 2000]
```

```python
combine = [df1900,df1950,df2000]

three_year = pd.concat(combine)
three_year
```

```python
three_year.plot(x = 'Year', y = ['Marriages_per_1000','Divorces_per_1000'], kind = 'bar')
```
